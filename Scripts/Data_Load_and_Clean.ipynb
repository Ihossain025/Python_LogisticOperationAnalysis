```python
# Fill the missing values in Dataset

# Merge only the columns we need from df_trips based on trip_id
# We use a 'left' join so we don't lose any rows from our fuel purchase table

df_fuel_pur = df_fuel_pur.merge(
    df_trips[['trip_id', 'truck_id', 'driver_id']], 
    on='trip_id', 
    how='left', 
    suffixes=('', '_from_trips')
)

# Use the pulled columns to fill the missing values in the original columns

df_fuel_pur['truck_id'] = df_fuel_pur['truck_id'].fillna(df_fuel_pur['truck_id_from_trips'])
df_fuel_pur['driver_id'] = df_fuel_pur['driver_id'].fillna(df_fuel_pur['driver_id_from_trips'])

# Drop the temporary columns we created during the merge

df_fuel_pur.drop(columns=['truck_id_from_trips', 'driver_id_from_trips'], inplace=True)

# Fill any absolute leftovers with 'Unknown'
df_fuel_pur['truck_id'] = df_fuel_pur['truck_id'].fillna('Unknown')
df_fuel_pur['driver_id'] = df_fuel_pur['driver_id'].fillna('Unknown')
```

```python
# Complete dictionary mapping including all unique origin and destination states
state_mapping = {
    'AZ': 'Arizona',        'CA': 'California',     'CO': 'Colorado', 
    'FL': 'Florida',        'GA': 'Georgia',        'IL': 'Illinois', 
    'IN': 'Indiana',        'MI': 'Michigan',       'MN': 'Minnesota', 
    'MO': 'Missouri',       'NC': 'North Carolina', 'NV': 'Nevada', 
    'NY': 'New York',       'OH': 'Ohio',           'OR': 'Oregon', 
    'PA': 'Pennsylvania',   'TN': 'Tennessee',      'TX': 'Texas', 
    'WA': 'Washington'
}

# Apply the mapping to both columns at the same time
state_cols = ['origin_state', 'destination_state']
df_routes[state_cols] = df_routes[state_cols].apply(lambda col: col.map(state_mapping))

# 3. Verify the changes
df_routes[state_cols].head()
```

```python
# This should print 0 if every single state abbreviation was successfully replaced

print("Missing values after mapping:", df_routes[state_cols].isnull().sum().sum())
```
